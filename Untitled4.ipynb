```
import matplotlib.pyplot as plt
import csv
f = open('/content/drive/My Drive/파이썬/다운로드받은거/ta.csv')
data = csv.reader(f)
next(data)
```

```
mx_temp = []
for row in data :
  if row[-1] != '' and row[0].split('-')[1] == '01' and row[0].split('-')[2] == '11':
    mx_temp.append(float(row[-1]))
plt.plot(mx_temp)
plt.show()
```

```
f = open('/content/drive/My Drive/파이썬/다운로드받은거/3e6d1b77a56b2f2b.csv')
data = csv.reader(f)
mytown = []
for row in data :
  if '부평동' in row[0] :
    for i in range(3, 104) :
      mytown.append(int(row[i]))
plt.figure(dpi=300)
plt.title('Bupyeong')
plt.plot(mytown)
plt.show
```
