```
prince = []
for line in file:
  line = line.strip()
  line = line.lower()
  words = line.split()
  prince.extend(words)
file.close() 
```

```
print(prince)
```

```
def count(d):
  freq = {}
  for i in d:
    if i not in freq:
      freq[i]=1
    elif i in freq:
      freq[i] +=1
  return freq
```

```
fr= {}
fr['학교']=1
fr
```

```
lp = count(prince)
lp
```

```
lp.items()
```

```
freq=sorted(lp.items(),key=lambda i:i[1],reverse=True)
freq=dict(freq)
freq
```

```
new_dic = {}
for key in freq:
  if len(key)>3:
    new_dic[key] = freq[key]
print(new_dic)
```

```
words = list(new_dic.keys())
words
```
